## Autofluorescence

```python
fname = os.path.join(FOLDER, '3Drecon-ADMBA-P4_atlasVolume.mhd')
itk_image = sitk.ReadImage(fname)
array = sitk.GetArrayFromImage(itk_image)
```

```python
array.shape
```

```python
print((array == 0).mean())
print((array < 0).mean())

sns.histplot(array[array>0].flatten(), log_scale=True);
```

```python
plt.imshow(array[:,200,:])
plt.show()
plt.imshow(array[190,:,:])
plt.show()
```
