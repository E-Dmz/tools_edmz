## Looking at lowest resolution (3)

### Channel s00: Nissl?

```python
cells = f['t00000']['s00']['3']['cells']
array = np.array(cells)
values, counts = np.unique(array, return_counts=True)
pd.DataFrame(dict(value=values, count=counts)).sort_values(by="count", ascending=False).reset_index(drop=True)
```

```python
print((array == 0).mean())
print((array < 0).mean())

sns.histplot(array[array>0].flatten(), log_scale=True);
```

```python
plt.imshow(array[:,50,:])
plt.show()
plt.imshow(array[71,:,:])
plt.show()
```
